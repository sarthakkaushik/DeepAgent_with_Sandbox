#### Sarthak Implemetaion for execute code tool

```python
from langchain.chat_models import init_chat_model
from deepagents import create_deep_agent
from e2b_code_interpreter import Sandbox


def execute_code(code):
    """Execute the given code in a sandbox environment and return the output."""
    if code:
        with Sandbox.create() as sandbox:
            execution = sandbox.run_code(code)
            return execution.text
    return None

agent = create_deep_agent()

system_prompt = "You are a research assistant.use \"execute_code\" tool to run the code in python and give me the answer in a deterministic way for any sort of calculation or python code execution."

agent = create_deep_agent(
    model=init_chat_model("openai:gpt-4o"),
    tools=[execute_code],

    system_prompt=system_prompt,
)

query = "Calculate how many r's are in the word 'strawberry'?"
query= "Analyze sample_data/expenses.csv and tell me the top 5 categories by total spend."

result = agent.invoke({"messages": [{"role": "user", "content": query}]})
# Print the agent's response
from IPython.display import Markdown, display
display(Markdown(result["messages"][-1].content))
```
